# Pengumpulan Data

Pada penelitian analisis sentimen ini, jenis data yang digunakan dalam penelitian ini adalah data tekstual berupa tweet yang sumber data dalam penelitian ini berasal dari platform Twitter dan bersifat data sekunder. Data diambil menggunakan API Twitter dengan menggunakan Tweet Harvest. Data diambil berdasarkan tanggal 22 Januari 2024 hingga 10 Februari 2024 dan berbahasa Indonesia. Tweet yang dikumpulkan disaring berdasarkan kata kunci "anies", "prabowo", "ganjar", "muhaimin", "gibran", dan "mahfud". Setelah data terkumpul, setiap tweet diberi label sentiment positif, negatif, atau netral secara manual. Data yang telah diberi label kemudian dibagi menjadi data training dan data testing dengan perbandingan 80:20.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from Modul._2_preprocessing_data import normalization
```

```python
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
df = pd.read_excel('Data/1_Pengumpulan_Data/tokoh_politik.xlsx', usecols=['full_text', 'sentiment'])
display(df)
```

```python
df['clean_full_text'] = df['full_text'].apply(lambda data: normalization(text=data, remove_punctuation_number=False))
df['word_count'] = df["clean_full_text"].str.split().str.len()
display(df)
display(df.describe())
display(df['sentiment'].value_counts())
```

```python
df_train, df_test = train_test_split(df, test_size=0.2, stratify=df['sentiment'], random_state=42)
df_train = pd.DataFrame(df_train).reset_index(drop=True)
df_test = pd.DataFrame(df_test).reset_index(drop=True)
```

```python
display(df_train)
display(df_train['sentiment'].value_counts())
```

```python
display(df_test)
display(df_test['sentiment'].value_counts())
```

```python
df_train.to_csv('Output/1_Pengumpulan_Data/train.csv', index=False)
df_test.to_csv('Output/1_Pengumpulan_Data/test.csv', index=False)
```
