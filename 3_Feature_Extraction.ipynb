```python
import pandas as pd
from Modul._2_preprocessing_data import cleaning_data
from Modul._3_feature_extraction import tfidf
```

```python
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
df_train = pd.read_csv('Output/2_Preprocessing_Data/train.csv', usecols=['text_preprocessed', 'sentiment'])
df_test = pd.read_csv('Output/2_Preprocessing_Data/test.csv', usecols=['text_preprocessed', 'sentiment'])
display(df_train)
display(df_test)
```

```python
display(df_train.shape)
display(df_train['text_preprocessed'].duplicated().sum())
display(df_train.isnull().sum())
df_train = cleaning_data(df_train, subset=['text_preprocessed'])
display(df_train.shape)
```

```python
display(df_test.shape)
display(df_test['text_preprocessed'].duplicated().sum())
display(df_test.isnull().sum())
df_test = cleaning_data(df_test, subset=['text_preprocessed'])
display(df_test.shape)
```

```python
tfidf_df_train = tfidf(df_train)
tfidf_df_test = tfidf(df_test)
display(tfidf_df_train)
display(tfidf_df_test)
```

```python
tfidf_df_train.to_parquet('Output/3_Feature_Extraction/train_tfidf.parquet', index=False)
tfidf_df_test.to_parquet('Output/3_Feature_Extraction/test_tfidf.parquet', index=False)
```

```python
df_train.to_csv('Output/3_Feature_Extraction/train.csv', index=False)
df_test.to_csv('Output/3_Feature_Extraction/test.csv', index=False)
```
