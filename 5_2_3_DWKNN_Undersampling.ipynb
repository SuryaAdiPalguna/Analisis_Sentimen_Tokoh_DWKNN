```python
import pandas as pd
import numpy as np
from Modul._5_implementasi_algoritma import DWKNN, undersampling
```

```python
import time
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
df_train = pd.read_csv('Output/3_Feature_Extraction/train.csv')
df_test = pd.read_csv('Output/3_Feature_Extraction/test.csv')
display(df_train)
display(df_test)
```

```python
tfidf_df_train = pd.read_parquet('Output/3_Feature_Extraction/train_tfidf.parquet')
tfidf_df_test = pd.read_parquet('Output/3_Feature_Extraction/test_tfidf.parquet')
display(tfidf_df_train)
display(tfidf_df_test)
```

```python
X_train = tfidf_df_train
X_test = tfidf_df_test
y_train = df_train['sentiment']
y_test = df_test['sentiment']
```

```python
X_train, y_train, X_test, y_test = undersampling(X_train, y_train, X_test, y_test)
```

```python
X = pd.concat([X_train, X_test], ignore_index=False).fillna(0).reset_index(drop=True)
X_train = X.loc[:len(X_train)-1].reset_index(drop=True)
X_test = X.loc[len(X_train):].reset_index(drop=True)
display(X_train)
display(X_test)
```

```python
X_train = np.array(X_train)
X_test = np.array(X_test)
y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
for i in range(10):
  start = time.time()
  result = pd.DataFrame()
  n = (i+1)*5
  dwknn = DWKNN(k=n)
  dwknn.fit(X_train, y_train)
  y_predict = dwknn.predict(X_test)
  result[f'k_{n}'] = pd.Series(y_predict)
  result.to_csv(f'Output/5_Implementasi_Algoritma/DWKNN/Undersampling/result_k_{n:02}.csv', index=False)
  end = time.time()
  print(f"Running time: {end - start:.4f} seconds")
```

```python
pd.DataFrame({'sentiment': y_test}).to_csv('Output/5_Implementasi_Algoritma/DWKNN/Undersampling/actual.csv', index=False)
```
