```python
import pandas as pd
from Modul._2_preprocessing_data import cleaning_data, case_folding, normalization, tokenizing, stopword_removal, stemming, remove_outlier
```

```python
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
df_train = pd.read_csv('Output/1_Pengumpulan_Data/train.csv', usecols=['full_text', 'sentiment', 'word_count'])
df_test = pd.read_csv('Output/1_Pengumpulan_Data/test.csv', usecols=['full_text', 'sentiment', 'word_count'])
display(df_train)
display(df_test)
```

### 1. Cleaning Data

```python
display(df_train.shape)
display(df_train['full_text'].duplicated().sum())
display(df_train.isnull().sum())
df_train = cleaning_data(df_train, subset=['full_text'])
display(df_train.shape)
```

```python
display(df_test.shape)
display(df_test['full_text'].duplicated().sum())
display(df_test.isnull().sum())
df_test = cleaning_data(df_test, subset=['full_text'])
display(df_test.shape)
```

### 2. Case Folding

```python
df_train['text_preprocessed'] = df_train['full_text'].apply(case_folding)
df_test['text_preprocessed'] = df_test['full_text'].apply(case_folding)
display(df_train)
display(df_test)
```

### 3. Normalization

```python
df_train['text_preprocessed'] = df_train['text_preprocessed'].apply(normalization)
df_test['text_preprocessed'] = df_test['text_preprocessed'].apply(normalization)
display(df_train)
display(df_test)
```

### 4. Tokenization

```python
df_train['text_preprocessed'] = df_train['text_preprocessed'].apply(tokenizing)
df_test['text_preprocessed'] = df_test['text_preprocessed'].apply(tokenizing)
display(df_train)
display(df_test)
```

### 5. Stopword Removal

```python
df_train['text_preprocessed'] = df_train['text_preprocessed'].apply(stopword_removal)
df_test['text_preprocessed'] = df_test['text_preprocessed'].apply(stopword_removal)
display(df_train)
display(df_test)
```

### 6. Stemming

```python
df_train['text_preprocessed'] = df_train['text_preprocessed'].apply(stemming)
df_test['text_preprocessed'] = df_test['text_preprocessed'].apply(stemming)
display(df_train)
display(df_test)
```

### 7. Remove Outlier

```python
df_train['text_preprocessed'] = df_train['text_preprocessed'].apply(lambda text: ' '.join(text))
df_test['text_preprocessed'] = df_test['text_preprocessed'].apply(lambda text: ' '.join(text))
display(df_train)
display(df_test)
```

```python
display(df_train.shape)
df_train = remove_outlier(df_train)
display(df_train.shape)
```

```python
display(df_test.shape)
df_test = remove_outlier(df_test)
display(df_test.shape)
```

### 8. Labelling Encoding

```python
df_train.loc[df_train['sentiment'] == 'positif', 'sentiment'] = df_test.loc[df_test['sentiment'] == 'positif', 'sentiment'] = 1
df_train.loc[df_train['sentiment'] == 'netral', 'sentiment'] = df_test.loc[df_test['sentiment'] == 'netral', 'sentiment'] = 0
df_train.loc[df_train['sentiment'] == 'negatif', 'sentiment'] = df_test.loc[df_test['sentiment'] == 'negatif', 'sentiment'] = -1
display(df_train)
display(df_test)
```

### Download

```python
df_train.to_csv('Output/2_Preprocessing_Data/train.csv', index=False)
df_test.to_csv('Output/2_Preprocessing_Data/test.csv', index=False)
```
